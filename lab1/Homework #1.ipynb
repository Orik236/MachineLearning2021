```python
tokens = [[i for i in re.split('[^a-z]', sentence) if i != ''] for sentence in sentences]
tokens
```

```python
unique_words = list(set([word for sentence in tokens for word in sentence]))
word2idx = {word : i for i, word in enumerate(unique_words)}
word2idx
```

```python
count_matrix = np.zeros((len(sentences), len(unique_words)))
count_matrix.shape
```

```python
for i, sentence in enumerate(tokens):
    for token in sentence:
        count_matrix[i, word2idx[token]] = sentence.count(token)
count_matrix
```

```python
def cos_distance(x, y):
    return 1 - x@y/(np.sqrt(x@x)*np.sqrt(y@y))
```

```python
distances = np.array([cos_distance(count_matrix[0], count_matrix[i]) for i in range(1, count_matrix.shape[0])])
distances
```

```python
nearest_sentences = np.argsort(distances)[:2] + 1
nearest_sentences
```

```python
with open('solution.txt', 'w') as f:
    f.write(f'{nearest_sentences[0]} {nearest_sentences[1]}')
```

### Better Method

```python
sentences = [re.sub(r'\d', ' ', sentence) for sentence in sentences]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_distances

vectorizer = CountVectorizer()
count_matrix = vectorizer.fit_transform(sentences)
count_matrix = count_matrix.toarray()
count_matrix.shape
```

```python
distances = cosine_distances(count_matrix[0].reshape(1, -1), count_matrix[1:])[0]
nearest_sentences = np.argsort(distances)[:2] + 1
nearest_sentences
```
